```python
def create_Vocab(wordsA, wordsB):
    
  vocab = set(wordsA).union(set(wordsB))
  Dict = {}

  Dict = dict.fromkeys(vocab, 0)
  cnt = 0

  for word in vocab:
    Dict[word] = cnt
    cnt += 1

  return vocab, Dict
```

```python
def computeTF(wordDict, words):
    tfDict = {}
    wordsCount = len(words)
    for word, count in wordDict.items():
        tfDict[word] = count/float(wordsCount)
    return tfDict
```

```python
import math
def computeIDF(docList):
    idfDict = {}
    N = len(docList)
    
    idfDict = dict.fromkeys(docList[0].keys(), 0)
    for doc in docList:
        for word, val in doc.items():
            if val > 0:
                idfDict[word] += 1
    
    for word, val in idfDict.items():
        idfDict[word] = math.log10(N / float(val))
        
    return idfDict
```

```python
def computeTFIDF(tfDocs, idfs):
    tfidf = {}
    for word, val in tfDocs.items():
        tfidf[word] = val*idfs[word]
    return tfidf
```

```python
import math
def cosSimilar(dA, dB): 
  N = len(dA)
  cos = 0.0
  tmp1 = 0.0
  tmp2 = 0.0
  for i in range(N):
    cos += dA[i]*dB[i]
    tmp1 += dA[i]*dA[i]
    tmp2 += dB[i]*dB[i]
  tmp1 = math.sqrt(tmp1)
  tmp2 = math.sqrt(tmp2)
  return cos/(tmp1*tmp2)
```

```python
import numpy as np
def create_Co_Matrix(docs, Dict, window_size=2): #doc = list
  N = len(Dict)
  coMatrix = np.full((N, N), 0)
  for doc in docs:
    print(doc)
    for i in range(len(doc)):
      word = doc[i]
      w = Dict[word]
      for step in range(1,window_size+1):
        if (i+step < len(doc)):
          u = doc[i+step]
          coMatrix[w][Dict[u]] += 1

        if (i-step >= 0):
          v = doc[i-step]
          coMatrix[w][Dict[v]] += 1
  return coMatrix
```
